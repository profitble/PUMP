```python
import requests
import json

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}

params = {
    "offset": 0,
    "limit": 100,
    "sort": "currently_live",
    "order": "DESC", 
    "includeNsfw": "true"
}

response = requests.get('https://frontend-api-v3.pump.fun/coins/currently-live',
                       params=params,
                       headers=headers)
coins = response.json()

# Save to file
with open('currently_live_coins.json', 'w') as f:
    json.dump(coins, f, indent=2)
```

```python
import random
import string

# Load initial tokens
with open('initial_tokens.json') as f:
    initial_tokens = json.load(f)

# Color options
backgrounds = ['#F59E0B', '#10B981', '#4F46E5', '#EC4899', '#6366F1', '#8B5CF6']
image_colors = ['#FFFBEB', '#ECFDF5', '#F8FAFC', '#FDF2F8', '#EEF2FF', '#F5F3FF'] 
text_colors = ['#FCD34D', '#34D399', '#93C5FD', '#F9A8D4', '#A5B4FC', '#C4B5FD']

# Generate random 4-char buyer names
def generate_buyer_name():
    return ''.join(random.choices(string.ascii_letters + string.digits, k=4))

# Generate 100 random trades
trades = []
buyer_names = set()
for i in range(100):
    token = random.choice(initial_tokens)
    color_idx = random.randint(0, len(backgrounds)-1)
    
    # Generate random amount avoiding clean numbers
    amount = random.uniform(0.1, 10.0)
    while amount % 0.5 == 0:
        amount = random.uniform(0.1, 10.0)
        
    # Generate unique buyer name
    buyer_name = generate_buyer_name()
    while buyer_name in buyer_names:
        buyer_name = generate_buyer_name()
    buyer_names.add(buyer_name)
        
    trade = {
        'id': str(i+1),
        'backgroundColor': backgrounds[color_idx],
        'imageColor': image_colors[color_idx], 
        'textDetailsColor': '#000000',
        'textColor': text_colors[color_idx],
        'buyerName': buyer_name,
        'amount': f"{amount:.1f}",
        'tokenName': token['token_symbol']
    }
    trades.append(trade)

# Save to file
with open('sample_trades.json', 'w') as f:
    json.dump(trades, f, indent=2)
```
